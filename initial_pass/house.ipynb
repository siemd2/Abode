```python
import numpy as np
import pandas as pd
import os
import math
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('Final_Updated_Housing.csv')
df.head()
```

```python
image_folder = 'Houses Dataset'

# Lists to hold the different types of images
bedroom_images = []
frontal_images = []
bathroom_images = []
kitchen_images = []

# Loop through the files in the specified directory
for filename in sorted(os.listdir(image_folder)):
    if '_bathroom' in filename:
        bathroom_images.append(filename)
    elif '_frontal' in filename:
        frontal_images.append(filename)
    elif '_bedroom' in filename:
        bedroom_images.append(filename)
    elif '_kitchen' in filename:
        kitchen_images.append(filename)
print(len(bedroom_images))
print(len(frontal_images))
# Now you have three lists with the categorized images
```

```python
df['front_image'] = frontal_images
df['bedroom_image']= bedroom_images
df['bathroom_image'] =bathroom_images
df['kitchen_image'] = kitchen_images
```

```python
df.head()

df.to_csv('image_housing.csv')
```

```python
inputs = np.array([300000, 2000, 3, 2, 2,2, 1, 0, 1, 1, 0, 0, 0.5])

headers = ['price','area','bedrooms','bathrooms','stories','mainroad',
           'guestroom','basement','hotwaterheating','airconditioning',
           'parking','prefarea','furnishingstatus','fake_address',
           'fake_zip_code'
]
for i in range(len(inputs)):
    if inputs[i] == 0:
        clean_data = df[df[headers[i]]>0]
    if inputs[i] == 1:
        clean_data = df[df[headers[i]]>=1]


```

```python
analog_inputs = inputs[:6]
analog_df = clean_data.iloc[:, 0:6]
print(analog_inputs)

analog_df.head()
```

```python
# Appending inputs to bottom row of data
analog_df = analog_df.append(analog_inputs, ignore_index=True)
```

```python
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(analog_df)
print(normalized_data)
```

```python
def percent_distance(point1, point2):
    return abs((point1-point2)/point1)

def euclidean_distance(point1, point2):
    point1 = point1
    point2 = point2
    return np.linalg.norm(point1 - point2)
```

```python
distances_list = []
prices_list = []
price_filtered = analog_df.iloc[:,0:1]
everything_else = analog_df.iloc[:,1:-1]
print(everything_else)

# Percent distances for price and area
for i in range(len(price_filtered)):
    percent = percent_distance(analog_inputs[0:1],np.array(price_filtered)[i])
    prices_list.append(percent)
print(prices_list)
# Euclidean distance for everything else
for i in range(len(normalized_data)):
    distance = euclidean_distance(analog_inputs,np.array(normalized_data)[i])
    distances_list.append(distance)

print(distances_list)
```

```python
# Concatenating prices_list and distances_list
full_list = pd.concat([pd.DataFrame(prices_list, columns=['prices']), distances_list], axis=1)
```

```python
for i in range(1, len(distances_list)):

    key = distances_list[i]

    # Move elements of arr[0..i-1], that are
    # greater than key, to one position ahead
    # of their current position
    j = i-1
    while j >= 0 and key < distances_list[j] :
            distances_list[j + 1] = distances_list[j]
            j -= 1
    distances_list[j + 1] = key

print(distances_list)
```
